```python
import numpy as np
with open('lplotneg.csv') as f:
    content = f.readlines()
lines = np.array(content) 
newlines = np.array([x.split(',') for x in lines])
index, x1, x2, x3 = [], [], [], []
for i in range(0, 100):
    index.append(newlines[i][0])
    x1.append(newlines[i][1])
    x2.append(newlines[i][2])
    x3.append(newlines[i][3])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.scatter(index, x1, marker='o', s=50, c='orange');
plt.scatter(index, x2, marker='o', s=50, c='blue');
plt.scatter(index, x3, marker='o', s=50, c='green');
plt.show()
```
